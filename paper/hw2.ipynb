```python
p, Deltax, ep= var('p, Deltax, ep')
q = 1 - p
H = p * log(1/p, 2) + (1-p) * log(1/(1-p), 2)
```

```python
# Problem 1a part 1
eq1 = limit(H(p + Deltax), Deltax=0, taylor=False)
view(eq1.taylor(p, 0, 6))
```

```python
# Problem 1a part 2
eq2 = limit(H(1/2 + ep + Deltax), Deltax=0)
# latex(eq2.taylor(ep, 0, 6))
view(eq2.taylor(ep, 0, 6))
```

```python
# Problem 1b
eq1 = H(p)
eq2 = p * log(1/p, 2)
eq3 = p * log(e/p, 2)

def make_label(l):
	return r"$%s$"%LatexExpr(l)
p1 = plot(eq1, (0, 1), color='red', legend_label=make_label('H(p)'))
p2 = plot(eq2, (0, 1), color='green', legend_label=make_label('p * log_2(1/p)'))
p3 = plot(eq3, (0, 1), color = 'blue', legend_label=make_label('p * log_2(e/p)'))
show(p1 + p2 + p3)
```
